```python
import pandas as pd
from glob import glob
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.DataFrame()
for file in glob('./traces/*.csv'):
    df = pd.concat([pd.read_csv(file),df],axis=1)
```

```python
for name in ['train_acc','train_loss','test_acc','test_loss']:
    df.filter(like=name).plot()
    plt.title(name)
    plt.savefig(f'./traces/{name}.png')
    plt.show()
```
